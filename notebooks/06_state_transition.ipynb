```python
import os
import pandas as pd
import numpy as np

STATE_PATH = "../outputs/tables/05D_financial_state_rule_labeled.csv"
OUT_DIR = "../outputs/tables"
os.makedirs(OUT_DIR, exist_ok=True)

df_state = pd.read_csv(STATE_PATH, encoding="utf-8-sig")

KEEP = [
    "Mã","Tên công ty",
    "Ngành ICB - cấp 1","Ngành ICB - cấp 2","Ngành ICB - cấp 3","Ngành ICB - cấp 4",
    "Năm","Financial_State_Rule","Pass_Ratio"
]
df_state = df_state[KEEP].copy()
df_state = df_state.sort_values(["Mã","Năm"]).reset_index(drop=True)

print("df_state:", df_state.shape)
df_state.head()
```

```python
STATE_ORDER = ["High_Risk","At_Risk","Stable","Healthy"]
state_rank = {s:i for i,s in enumerate(STATE_ORDER)}

def trend_label(g):
    ranks = [state_rank.get(x, np.nan) for x in g["Financial_State_Rule"].tolist()]
    if any(pd.isna(r) for r in ranks) or len(ranks) < 2:
        return np.nan
    delta = ranks[-1] - ranks[0]
    if delta > 0:
        return "Improving"
    if delta < 0:
        return "Deteriorating"
    return "Stable"

trend_df = (
    df_state
    .groupby(["Mã","Tên công ty","Ngành ICB - cấp 1"], as_index=False)
    .apply(lambda g: pd.Series({
        "Trend": trend_label(g.sort_values("Năm")),
        "Year_Min": int(g["Năm"].min()),
        "Year_Max": int(g["Năm"].max()),
        "Num_Years": int(g["Năm"].nunique())
    }))
    .reset_index(drop=True)
)

trend_summary = (
    trend_df["Trend"]
    .value_counts(dropna=False)
    .rename_axis("Trend")
    .reset_index(name="Count")
)
trend_summary["Share"] = (trend_summary["Count"] / trend_summary["Count"].sum()).round(3)

trend_path = os.path.join(OUT_DIR, "06D1_trend_summary.csv")
trend_summary.to_csv(trend_path, index=False, encoding="utf-8-sig")
print("✅ Saved:", trend_path)

trend_summary
```

```python
df_state2 = df_state.sort_values(["Mã","Năm"]).copy()

df_state2["State_Next"] = df_state2.groupby("Mã")["Financial_State_Rule"].shift(-1)
df_state2["Year_Next"]  = df_state2.groupby("Mã")["Năm"].shift(-1)

trans = df_state2[
    df_state2["State_Next"].notna() &
    (df_state2["Year_Next"] == df_state2["Năm"] + 1)
].copy()

trans["YearPair"] = trans["Năm"].astype(int).astype(str) + "->" + trans["Year_Next"].astype(int).astype(str)

print("trans:", trans.shape)
trans.head()
```

```python
STATE_ORDER = ["High_Risk","At_Risk","Stable","Healthy"]

def transition_matrix_for(df_sub):
    count = pd.crosstab(df_sub["Financial_State_Rule"], df_sub["State_Next"]).reindex(
        index=STATE_ORDER, columns=STATE_ORDER
    )
    prob = count.div(count.sum(axis=1), axis=0)
    return count, prob

yearpair_stats = []
for yp, g in trans.groupby("YearPair"):
    count, prob = transition_matrix_for(g)
    # lưu dạng long cho dễ dashboard
    for i in STATE_ORDER:
        for j in STATE_ORDER:
            yearpair_stats.append({
                "YearPair": yp,
                "From": i,
                "To": j,
                "Count": float(count.loc[i, j]) if pd.notna(count.loc[i, j]) else 0.0,
                "Prob": float(prob.loc[i, j]) if pd.notna(prob.loc[i, j]) else np.nan
            })

yearpair_df = pd.DataFrame(yearpair_stats)
out_path = os.path.join(OUT_DIR, "06D2_yearpair_transition_long.csv")
yearpair_df.to_csv(out_path, index=False, encoding="utf-8-sig")
print("✅ Saved:", out_path)

yearpair_df.head(12)
```

```python
import plotly.graph_objects as go

def sankey_yearpair(df_trans, yearpair, industry=None, min_count=5):
    """
    df_trans: bảng trans (records) có cột YearPair, Financial_State_Rule, State_Next, Ngành ICB - cấp 1
    yearpair: "2020->2021" ...
    industry: None hoặc tên ngành cấp 1
    min_count: lọc luồng nhỏ
    """
    d = df_trans[df_trans["YearPair"] == yearpair].copy()
    if industry is not None:
        d = d[d["Ngành ICB - cấp 1"] == industry].copy()

    if d.empty:
        raise ValueError("Không có dữ liệu cho bộ lọc đã chọn.")

    flow = (
        d.groupby(["Financial_State_Rule","State_Next"])
        .size()
        .reset_index(name="Count")
    )
    flow = flow[flow["Count"] >= min_count].copy()

    states_left  = [f"{yearpair.split('->')[0]}:{s}" for s in STATE_ORDER]
    states_right = [f"{yearpair.split('->')[1]}:{s}" for s in STATE_ORDER]
    nodes = states_left + states_right
    node_index = {n:i for i,n in enumerate(nodes)}

    flow["Source"] = flow["Financial_State_Rule"].apply(lambda s: node_index[f"{yearpair.split('->')[0]}:{s}"])
    flow["Target"] = flow["State_Next"].apply(lambda s: node_index[f"{yearpair.split('->')[1]}:{s}"])

    title = f"Sankey {yearpair}"
    if industry is not None:
        title += f" | {industry}"

    fig = go.Figure(
        data=[go.Sankey(
            node=dict(label=nodes),
            link=dict(
                source=flow["Source"].tolist(),
                target=flow["Target"].tolist(),
                value=flow["Count"].tolist()
            )
        )]
    )
    fig.update_layout(title_text=title, font_size=12)
    return fig
```

```python
# chọn cặp năm đầu tiên có dữ liệu
sample_yearpair = trans["YearPair"].value_counts().index[0]
print("Sample yearpair:", sample_yearpair)

fig = sankey_yearpair(trans, sample_yearpair, industry=None, min_count=10)
fig.show()
```

```python
# =========================================================
# (6) Table 5-style output: Firm-level adjacent-year transitions
#     (Ticker, Company, Industry, Year(t), Year(t+1), State(t), State(t+1))
# =========================================================
import os
import pandas as pd

STATE_PATH = "../outputs/tables/05D_financial_state_rule_labeled.csv"
OUT_DIR = "../outputs/tables"
os.makedirs(OUT_DIR, exist_ok=True)

df = pd.read_csv(STATE_PATH, encoding="utf-8-sig")

# ---- Choose ONE industry column for the table (Level-1 as in your example)
IND_COL = "Ngành ICB - cấp 1"

# ---- Keep needed columns
keep = ["Mã", "Tên công ty", IND_COL, "Năm", "Financial_State_Rule"]
df = df[keep].copy().sort_values(["Mã", "Năm"]).reset_index(drop=True)

# ---- Build adjacent-year transitions
df["Year(t+1)"] = df.groupby("Mã")["Năm"].shift(-1)
df["Financial State (t+1)"] = df.groupby("Mã")["Financial_State_Rule"].shift(-1)

df_trans = df[
    df["Financial State (t+1)"].notna() &
    (df["Year(t+1)"] == df["Năm"] + 1)
].copy()

# ---- Rename columns to match Table 5 style
df_trans = df_trans.rename(columns={
    "Mã": "Ticker",
    "Tên công ty": "Company Name",
    IND_COL: "Industry",
    "Năm": "Year (t)",
    "Financial_State_Rule": "Financial State (t)"
})

df_trans["Year (t+1)"] = df_trans["Year(t+1)"].astype(int)
df_trans = df_trans.drop(columns=["Year(t+1)"])

# ---- Reorder columns
table5_cols = [
    "Ticker",
    "Company Name",
    "Industry",
    "Year (t)",
    "Year (t+1)",
    "Financial State (t)",
    "Financial State (t+1)"
]
df_table5 = df_trans[table5_cols].copy()

# ---- Save
out_path = os.path.join(OUT_DIR, "06D4_table5_adjacent_year_transitions.csv")
df_table5.to_csv(out_path, index=False, encoding="utf-8-sig")
print("✅ Saved:", out_path)

# ---- Preview (optional)
df_table5.head(20)
```

```python
# =============================
# TABLE 6: Corporate financial state during 2020–2024
# =============================
import os
import numpy as np
import pandas as pd

OUT_DIR = "../outputs/tables"
os.makedirs(OUT_DIR, exist_ok=True)

STATE_ORDER = ["High_Risk", "At_Risk", "Stable", "Healthy"]
state_rank = {s: i for i, s in enumerate(STATE_ORDER)}

def overall_state_from_mean_pass_ratio(x):
    if pd.isna(x):
        return np.nan
    if x < 0.25:
        return "At Risk" if False else "High Risk"   # giữ nguyên logic của bạn
    if x < 0.50:
        return "At Risk"
    if x < 0.75:
        return "Stable"
    return "Healthy"

def trend_from_states(states):
    ranks = [state_rank.get(s, np.nan) for s in states]
    if any(pd.isna(r) for r in ranks) or len(ranks) < 2:
        return np.nan
    delta = ranks[-1] - ranks[0]
    if delta > 0:
        return "Improving"
    if delta < 0:
        return "Deteriorating"
    return "Stable"

# IMPORTANT: dùng df_state đã có sẵn từ đoạn trên (đã gồm Pass_Ratio + Financial_State_Rule)
# df_state = ... (đang có ở notebook của bạn)

rows = []
group_cols = ["Mã", "Tên công ty", "Ngành ICB - cấp 1"]

for (ticker, name, industry), g in df_state.groupby(group_cols):
    g = g.sort_values("Năm")

    start_year = int(g["Năm"].min())
    end_year   = int(g["Năm"].max())

    mean_pass = pd.to_numeric(g["Pass_Ratio"], errors="coerce").mean(skipna=True)
    overall_state = overall_state_from_mean_pass_ratio(mean_pass)

    states = g["Financial_State_Rule"].tolist()
    trend = trend_from_states(states)

    rows.append({
        "Symbol": ticker,
        "Company Name": name,
        "Industry": industry,
        "Start Year": start_year,
        "End Year": end_year,
        "Overall State": overall_state,
        "Trend": trend
    })

table6 = pd.DataFrame(rows)

# ====== SORT OPTIONS (chọn 1) ======
# Option A: Sort toàn bảng theo Symbol (giống kiểu “bảng danh sách”)
table6 = table6.sort_values(["Symbol"]).reset_index(drop=True)

# Option B: Nếu bạn muốn group theo Industry nhưng vẫn giữ Symbol tăng dần trong mỗi ngành
# table6 = table6.sort_values(["Industry", "Symbol"]).reset_index(drop=True)

# ===================================
table6_path = os.path.join(OUT_DIR, "06B_table6_corporate_financial_state_2020_2024.csv")
table6.to_csv(table6_path, index=False, encoding="utf-8-sig")
print("✅ Saved Table 6:", table6_path)

table6.head(20)
```
