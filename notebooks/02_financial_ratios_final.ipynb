```python
import os
import numpy as np
import pandas as pd

RATIOS_PATH = "../data/processed/financial_ratios_2020_2024.csv"
OUT_DIR = "../outputs/tables"
os.makedirs(OUT_DIR, exist_ok=True)

df = pd.read_csv(RATIOS_PATH, encoding="utf-8-sig")
print("Loaded:", df.shape)
df.head(3)
```

```python
missing_rate = df.isna().mean().sort_values(ascending=False)
print("Top 20 missing columns:")
display(missing_rate.head(20))
```

```python
# =========================
# Missing analysis theo NGÀNH (ICB cấp 1)
# =========================

ratio_cols = [
    "Current_Ratio", "Quick_Ratio", "Cash_Ratio",
    "Debt_Equity", "Net_Leverage",
    "Asset_Turnover", "Fixed_Asset_Turnover",
    "Inventory_Turnover", "Receivables_Turnover",
    "ln_Total_Assets", "ln_Revenue",
    "ROA", "ROE", "ROS"
]

missing_by_industry = (
    df
    .groupby("Ngành ICB - cấp 1")[ratio_cols]
    .apply(lambda x: x.isna().mean())
    .round(3)
)

missing_by_industry
```

```python
# Các ngành tài chính cần loại bỏ
FINANCIAL_INDUSTRIES = [
    "Ngân hàng",
    "Tài chính",
    "Không xác định"
]
print("Số DN theo ngành:")
display(
    df["Ngành ICB - cấp 1"]
    .value_counts()
)
```

```python
df_clean = (
    df
    .loc[~df["Ngành ICB - cấp 1"].isin(FINANCIAL_INDUSTRIES)]
    .copy()
)

print("Sau khi loại ngành ngân hàng, tài chính & không xác định :")
print("Shape:", df_clean.shape)
print("Ngành còn lại:", df_clean ["Ngành ICB - cấp 1"].unique())
```

```python
missing_rate_clean = (
    df_clean
    .isna()
    .mean()
    .sort_values(ascending=False)
)

print("Top missing columns (đã clean):")
display(missing_rate_clean.head(20))
```

```python
ratio_cols = [
    "Current_Ratio", "Quick_Ratio", "Cash_Ratio",
    "Debt_Equity", "Net_Leverage",
    "Asset_Turnover", "Fixed_Asset_Turnover",
    "Inventory_Turnover", "Receivables_Turnover",
    "ROA", "ROE", "ROS",
    "ln_Total_Assets", "ln_Revenue"
]

missing_by_industry_clean = (
    df_clean
    .groupby("Ngành ICB - cấp 1")[ratio_cols]
    .apply(lambda x: x.isna().mean())
    .round(3)
)

display(missing_by_industry_clean)
```

```python
FINAL_RATIO_COLS = [
    # Thanh khoản
    "Current_Ratio", "Quick_Ratio", "Cash_Ratio",

    # Đòn bẩy
    "Debt_Equity", "Net_Leverage",

    # Hiệu quả
    "Asset_Turnover", "Fixed_Asset_Turnover",

    # Sinh lợi
    "ROA", "ROE", "ROS",

    # Quy mô (phân nhóm, không scoring)
    "ln_Total_Assets", "ln_Revenue"
]

META_COLS = [
    "Mã", "Tên công ty", "Sàn",
    "Ngành ICB - cấp 1", "Ngành ICB - cấp 2",
    "Ngành ICB - cấp 3", "Ngành ICB - cấp 4",
    "Năm"
]
```

```python
df_final = (
    df_clean
    [META_COLS + FINAL_RATIO_COLS]
    .copy()
)

print("Final shape:", df_final.shape)
df_final.head()
```

```python
final_missing = df_final.isna().mean().sort_values(ascending=False)
display(final_missing)
```

```python
FINAL_PATH = "../data/processed/financial_ratios_final_clean.csv"

df_final.to_csv(FINAL_PATH, index=False, encoding="utf-8-sig")

print("Saved final clean dataset to:", FINAL_PATH)
```
